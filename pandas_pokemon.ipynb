```python
## Filter out all the names that contain 'Mega'

df.loc[df['Name'].str.contains('Mega')]
```

```python
## Within the .loc function we can use the ~ to remove 

df.loc[~df['Name'].str.contains('Mega')] # you can also pass different functions in this as well
```

```python
import re #regular expressions; helps detect expressions in data

df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)] #case sensitive
```

```python
## To ignore any case sensitive text use flags=re.I

df.loc[df['Type 1'].str.contains('Fire|Grass', flags=re.I, regex=True)]
```

```python
## Using regex=True -- If you want to just get text that start with 'Pi' you can do the following
## in str.contains('pi[a-z]*, flags=re.I, regex=True) "*" means Zero or more
df.loc[df['Name'].str.contains('^pi[a-z]', flags=re.I, regex=True)] 
```

Aggregate Statistics (Group By)

```python
## Load csv

df = pd.read_csv('modified.csv')

df
```

```python
## Finding average HP & Attack grouped by Type 1
df = pd.read_csv('modified.csv')

df.groupby('Type 1').mean(numeric_only=True) #this lets pandas pick numeric columns automatically OR
```

```python
## Using the sort_values function 

df.groupby('Type 1').mean(numeric_only=True).sort_values('Defense', ascending=False) # ascending=False to see the highest defense
```

```python
# Find highest attack

df.groupby('Type 1').mean(numeric_only=True).sort_values('Attack', ascending=False) # ascending=False to see the highest defense
```

```python
## When using .groupby you can use .mean(), .sum(), .count()

df['count'] = 1

df.groupby(['Type 1']).count() # displays the number of each Type 1 and counts them 
```

```python
## To just get the count column 

df.groupby(['Type 1']).count()['count']
```

```python
## Group by Multiple Parameters

df.groupby(['Type 1', 'Type 2']).count()['count'] # useful when working with a large dataset if you can't load into one dataframe
```

Working with large amounts of data

```python
## Reading in chunk of data instead of whole dataset

# regular way to load in ALL data 

# df = pd.read_csv('modified.csv')

## Loading in chunks of data

for df in pd.read_csv('modified.csv', chunksize=5):
    print("CHUNK DF")
    print(df)


```

```python
## You can also define a new dataframe

new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv',chunksize=5):
    results = df.groupby(['Type 1']).count() #counts the Type 1 amount 

    new_df = pd.concat([new_df, results])

new_df
```
